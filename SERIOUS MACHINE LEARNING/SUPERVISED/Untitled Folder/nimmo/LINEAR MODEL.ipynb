# importing the iris data

```python
dir(mglearn.datasets)
dir(sklearn.datasets)
from sklearn.datasets import load_iris
data=load_iris()
print(data)
```

getting the keys

```python
data.keys()
```

```python
print(data.DESCR[:250] +"\n...")
print(data.feature_names)
print(data.target_names)
print(data.target) #depoendent variable ([]}
print(data.data)
```

# for analysis multidimensional array(has [[]]) are independent variables(has []) that is x .while singulkar arrays are dependent variables

```python
x=data.data
print(x)
y=data.target
print(y)
```

```python
dir(sklearn)
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print('the R2 of train  is  :{:.2f}'.format(lr.score(x_train,y_train)))
```

```python
print("the R2 of test is :{:.2f}".format(lr.score(x_test,y_test)))
```

```python
print("The slope is:{}".format(lr.coef_))
```

```python
print("The intercept is :{}".format(lr.intercept_))
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
rg=Ridge()
rg.fit(x_train,y_train)
```

```python
print("The R2 of train is :{:.2f}".format(rg.score(x_train,y_train)))
```

```python
print("The R2 of test is :{:.2f}".format(rg.score(x_test,y_test)))
```

```python
print("The slope is :{}".format(rg.coef_))
```

```python
print("The slope is :{}".format(rg.intercept_))
```
