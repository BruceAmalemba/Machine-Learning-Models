# underfitting

It happens when a model is too simple to capture the patterns in the dataâ€”like trying to fit a straight line through a curvy scatter plot.

```python
import mglearn
dir(mglearn)
dir(mglearn.datasets)
```

# importing make_forge

```python
from mglearn.datasets import make_forge
X,y=make_forge()
print(X,y)
```

multi dimensional arrays are always the independent variables (X) while the singular dimensional array is the 
dependent variable (y)

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
model=LinearRegression()
model.fit(X_train,y_train)
```

# testing if the model is good we use the R SCORE

Rand score (also known as the Rand index), a common metric used in clustering to measure the similarity between two sets of cluster labels.

```python
print("the R2 of train is:{:.2f}".format(model.score(X_train,y_train)))
```

```python
print("the R2 of test is:{:.2f}".format(model.score(X_test,y_test)))
```

checking for the slope 

```python
print("The slope is :{}".format(model.coef_))
```
