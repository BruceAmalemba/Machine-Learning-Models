```python
import sklearn
import mglearn
dir(mglearn.datasets)
from mglearn.datasets import make_forge
X,y=make_forge()
print(X,y)
```

```python
dir(sklearn.neighbors)
from sklearn.neighbors import KNeighborsClassifier
KNC=KNeighborsClassifier()
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
KNC.fit(X_train,y_train)
```

# getting the rand score

```python
print("The rand score of test is :{:.2f}".format(KNC.score(X_test,y_test)))
```

```python
print("The rand score of train is :{:.2f}".format(KNC.score(X_train,y_train)))
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train,y_train)
```

```python
print("The rand score of test is :{:.2f}".format(linreg.score(X_test,y_test)))
```

```python
print("The rand score of train is :{:.2f}".format(linreg.score(X_train,y_train)))
```

```python
from mglearn.datasets import make_forge
X,y=make_forge()
print(X,y)
from sklearn.neighbors import KNeighborsClassifier
KNC_1=KNeighborsClassifier(n_neighbors=3)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=0)
KNC_1.fit(X_train,y_train)
```

```python
print("The R2 of train is :{:.2f}".format(KNC_1.score(X_train,y_train)))
```

```python
print("The R2 of test is :{:.2f}".format(KNC_1.score(X_test,y_test)))
```

```python
KNC_2=KNeighborsClassifier(n_neighbors=9)
KNC_2.fit(X_train,y_train)
```

```python
print("The R2 of test is :{:.2f}".format(KNC_2.score(X_test,y_test)))
```

```python
print("The R2 of train is :{:.2f}".format(KNC_2.score(X_train,y_train)))
```
