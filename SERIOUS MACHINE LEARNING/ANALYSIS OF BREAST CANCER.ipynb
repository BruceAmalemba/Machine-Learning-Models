# TRAINING AND TESTING

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
lr=LinearRegression()
lr.fit(x_train,y_train)
```

checking for the rand score

```python
print("The rand score of train is :{:.2f}".format(lr.score(x_train,y_train)))
```

```python
print("The rand score of test is :{:.2f}".format(lr.score(x_test,y_test)))
```

the variable names are affecting the coeficients.since we had 30 variable names we have 30 coef_

```python
print("The  coeffients are :{}".format(lr.coef_))
```

```python
print("The  intercept is :{}".format(lr.intercept_))
```

```python
print(f" y={lr.coef_}x+{lr.intercept_}")
```

# RIDGE REGRESSION
ridge increases the rand score(r squared of testing) or reduces the slope/gradient ie converges the gradient to 0 
hence inreasing the accuracy

```python
dir(sklearn.linear_model)
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
rr=Ridge()
rr.fit(x_train,y_train)
```

```python
print("The rand score of train is :{:.2f}".format(rr.score(x_train,y_train)))
```

```python
print("The rand score of test is :{:.2f}".format(rr.score(x_test,y_test)))
```

```python
print("The  coeffients are :{}".format(rr.coef_))
```

```python
print("The  intercept is :{}".format(rr.intercept_))
```

```python
print(f" y={rr.coef_}x+{rr.intercept_}")
```

# deacreasing alpha and increasing the number of iterations increases the level of accuracy
the default alpha=1

```python
from sklearn.linear_model import Lasso
model1=Lasso(alpha=0.1,max_iter=10000)
model1.fit(x_train,y_train)
```

```python
print("The rand score of test is :{:.2f}".format(model1.score(x_test,y_test)))
```

```python
model2=Lasso(alpha=0.01,max_iter=10000)
model2.fit(x_train,y_train)
```

```python
from  sklearn.linear_model import Ridge
ridg=Ridge(alpha=0.1)
ridg.fit(x_train,y_train)
```
